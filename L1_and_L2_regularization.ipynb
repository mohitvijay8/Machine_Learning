```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Melbourne_housing.csv')
```

```python
df.shape
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
cols_useful = ['Suburb','Type','Rooms','Method','SellerG','Regionname','Propertycount','Distance',
              'CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
```

```python
dataset = df[cols_useful]
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset.shape
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
```

```python
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
x = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
```

```python
las = Lasso(alpha=50,max_iter=100,tol=0.1)
```

```python
las.fit(x_train,y_train)
```

```python
las.score(x_test,y_test)
```

```python
las.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
```

```python
rigmodel = Ridge(alpha=50,max_iter=100,tol=0.1)
```

```python
rigmodel.fit(x_train,y_train)
```

```python
rigmodel.score(x_test,y_test)
```

```python
rigmodel.score(x_train,y_train)
```
